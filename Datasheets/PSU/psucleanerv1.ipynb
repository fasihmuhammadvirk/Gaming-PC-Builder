```python
import pandas as pd
```

```python
df = pd.read_csv('powersupply.csv')
df
```

```python
df['name'] = df['name'].str.replace(r'\s*\(\d{4}\)', '', regex=True)
df.drop_duplicates(subset=['name', 'wattage', 'efficiency'], inplace=True)
df
```

```python
efficiency_mapping = {
    'plus': 80,
    'bronze': 82,
    'silver': 85,
    'gold': 87,
    'platinum': 90,
    'titanium': 94,
    '': 0
}
df['efficiency_type'] = df['efficiency']
df['efficiency_type'] = df['efficiency_type'].fillna('white')
df['efficiency'] = df['efficiency'].map(efficiency_mapping).fillna(0)
df
```

```python
modular_mapping = {
    'Full': 10,
    'Semi': 6,
    'false': 0
}
df['modularity'] = df['modular']
df['modular'] = df['modular'].map(modular_mapping).fillna(0)
df
```

```python
df['power'] = df['wattage']
df
```

```python
df.to_csv('psu_v1.csv', index=False)
```

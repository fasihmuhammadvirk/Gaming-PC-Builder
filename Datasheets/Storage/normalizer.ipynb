```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Projects\0FYP\Datasheets\Storage\storage.csv')
df.head()
```

```python
#'storage': {'capacity': 0.4, 'speed': 0.3, 'price_per_gb': 0.2, 'cache': 0.1},
```

```python
df[['capacity_spec', 'cache_spec', 'speed_spec']] = df[['capacity', 'cache', 'speed']]
df['capacity_spec'] = df['capacity_spec'] / 1000
df
```

```python
#'gpu': {'memory': 2.2, 'memory_ver': 1, 'g3dmark': 1.2, 'api_score': 1.2, 'blender_score': 1.2, 'fps_score': 1.2},
scaler_min_max = MinMaxScaler()
df[['capacity', 'cache', 'speed', 'price_per_gb']] = scaler_min_max.fit_transform(df[['capacity', 'cache', 'speed', 'price_per_gb']])
df
```

```python
#'gpu': {'memory': 2.2, 'memory_ver': 1, 'g3dmark': 1.2, 'api_score': 1.2, 'blender_score': 1.2, 'fps_score': 1.2},
```

```python
df['price_per_gb'] = 1 - df['price_per_gb']
df
```

```python
df['type'] = df['type'].str.replace('7200', 'HDD')
df['type'] = df['type'].str.replace('5400', 'HDD')
df['type'] = df['type'].str.replace('5900', 'HDD')
df['type'] = df['type'].str.replace('10000', 'HDD')
df['type'] = df['type'].str.replace('5700', 'HDD')
df['type'] = df['type'].str.replace('5640', 'HDD')
df['type'] = df['type'].str.replace('5000', 'HDD')
df['type'] = df['type'].str.replace('5200', 'HDD')
df['type'] = df['type'].str.replace('5760', 'HDD')
df['type'] = df['type'].str.replace('5940', 'HDD')
df
```

```python
df.to_csv('storage_final.csv', index=False)
```

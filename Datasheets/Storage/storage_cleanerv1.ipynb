```python
import pandas as pd
```

```python
df = pd.read_csv('harddrive.csv')
df
```

```python
df['form_factor'] = df['form_factor'].str.replace('-2280', '')
df['cache'] = df['cache'].fillna(0)
df
```

```python
filtered_interfaces = {
    'SATA 6.0 Gb/s': 550,
    'SATA 3.0 Gb/s': 300,
    'M.2 PCIe 3.0 X4': 3500,
    'M.2 PCIe 4.0 X4': 7000,
    'M.2 SATA': 550
}

# Filter the DataFrame to keep only the valid interfaces
df = df[df['interface'].isin(filtered_interfaces.keys())]

# Create the 'bandwidth' column and fill it with the corresponding values
df['speed'] = df['interface'].map(filtered_interfaces)
df.loc[df['type'].isin(['7200', '5400']), 'speed'] = 160

df
```

```python
df.to_csv('storage.csv', index=False)
```

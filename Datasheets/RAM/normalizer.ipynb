```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Projects\0FYP\Datasheets\RAM\ram.csv')
df.head()

#{'total_capacity_gb': 1, 'first_word_latency': 0.1, 'cas_latency': 0.2, 'speed_mhz': 0.4, 'price_per_gb':0.3}
```

```python
df = df.drop(columns=['speed', 'modules'])
```

```python
df[['capacity', 'latency', 'speed']] = df[['total_capacity_gb', 'cas_latency', 'speed_mhz']]

df
```

```python
scaler_min_max = MinMaxScaler()
df[['total_capacity_gb', 'first_word_latency', 'cas_latency', 'speed_mhz', 'price_per_gb']] = scaler_min_max.fit_transform(df[['total_capacity_gb', 'first_word_latency', 'cas_latency', 'speed_mhz', 'price_per_gb']])
df
```

```python
df['price_per_gb'] = 1 - df['price_per_gb']
df['first_word_latency'] = 1 - df['first_word_latency']
df['cas_latency'] = 1 - df['cas_latency']
df
```

```python
df.to_csv('memory_final.csv', index=False)
```

```python
import pandas as pd
import pickle
import numpy as np
```

```python
imp_cpu_pass = pd.read_csv('IMP_cpu_pass_cine.csv')
imp_cpu_pass.head()
```

```python
blender = pd.read_csv(r'D:\Projects\0FYP\Datasheets\CPU\blender - cleaned.csv')
blender = blender.drop(columns=['Unnamed: 0'])
blender
```

```python
imp_cpu_pass_cine = pd.merge(imp_cpu_pass, blender, on='name', how='left')
```

```python
with open('PDblender.pkl', 'rb') as file:
    model = pickle.load(file)

input_columns = ['core_count', 'boost_clock', 'tdp', 'smt', 'cpuMark']

df_new = imp_cpu_pass_cine
for index, row in df_new.iterrows():
    if np.isnan(row['Blender Score']):
        X_pred = row[input_columns].values.reshape(1, -1)
        y_pred = model.predict(X_pred)      # making the prediction
        df_new.at[index, 'Blender Score'] = int(y_pred)
        df_new.at[index, 'Number of Benchmarks'] = 0
    # else:
        # df_new.at[index, 'cinePred'] = 'FALSE'
```

```python
df_new.head()
```

```python
count_predicted = (df_new['Number of Benchmarks'] == 0).sum()
count_real = (df_new['Number of Benchmarks'] > 0).sum()

print(f'predicted: {count_predicted}, real: {count_real}, total: {count_real + count_predicted}')
```

```python
df_new.to_csv('IMP_cpu_pass_cine_blender.csv', index=False)
```

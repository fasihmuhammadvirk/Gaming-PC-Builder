```python
import pandas as pd
import pickle
import numpy as np
```

```python
imp_cpu_pass = pd.read_csv('IMP_cpu_pass.csv')
imp_cpu_pass = imp_cpu_pass.drop(columns=['threadMark'])

imp_cpu_pass.head()
```

```python
cinebench = pd.read_csv(r'D:\Projects\0FYP\Datasheets\CPU\cinebench_r15.csv')
cinebench = cinebench.rename(columns={'match': 'name'})
cinebench = cinebench.drop(columns={'Cores', 'GHz'})
cinebench
```

```python
imp_cpu_pass = pd.merge(imp_cpu_pass, cinebench, on='name', how='left')
```

```python
with open('PDcinebench.pkl', 'rb') as file:
    model = pickle.load(file)

input_columns = ['core_count', 'boost_clock', 'tdp', 'smt', 'cpuMark']

df_new = imp_cpu_pass
for index, row in df_new.iterrows():
    if np.isnan(row['CPU ScoreMulti 15']):
        X_pred = row[input_columns].values.reshape(1, -1)
        y_pred = model.predict(X_pred)      # making the prediction
        df_new.at[index, 'CPU ScoreMulti 15'] = int(y_pred)
        df_new.at[index, 'cinePred'] = 'TRUE'
    else:
        df_new.at[index, 'cinePred'] = 'FALSE'
```

```python
df_new = df_new.drop(columns=['CPU ScoreSingle 15'])
df_new.head()
```

```python
count_predicted = (df_new['cinePred'] == 'TRUE').sum()
count_real = (df_new['cinePred'] == 'FALSE').sum()

print(f'predicted: {count_predicted}, real: {count_real}, total: {count_real + count_predicted}')
```

```python
df_new.to_csv('IMP_cpu_pass_cine.csv', index=False)
```

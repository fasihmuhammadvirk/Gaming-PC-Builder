```python
import pandas as pd
```

```python
df = pd.read_csv('IMP_cpu_pass_cine_blender.csv')
```

```python
df['price'] = df['price'].fillna(df['pmprice'])
df = df.drop(columns={'pmprice'})
df = df.rename(columns={'cpuMark': 'passmark', 'CPU ScoreMulti 15': 'cinebench', 'Blender Score': 'blender', 'Number of Benchmarks': 'blenderPred'})
df['category'] = df['category'].fillna('Unknown')
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df_normalized = df
columns_to_normalize = ['passmark', 'cinebench', 'blender']

scaler = MinMaxScaler()
df_normalized[columns_to_normalize] = scaler.fit_transform(df_normalized[columns_to_normalize])
df_normalized
```

```python
remap = {
    'AMD 300 series': 'A320, B350, X370',
    'AMD 300 Series': 'A320, B350, X370',
    'AMD 300 Series*': 'A320, B350, X370',
    'AMD 400 Series': 'B450, X470',
    'AMD 500 Series': 'A520, B550, X570',
    'Intel 100 Series': 'H110, B150, H170',
    'Intel 200 Series': 'B250, H270, Z270',
    'Intel 300 Series': 'H310, B360, B365, H370, Z390',
    'Intel 400 Series': 'H410, B460, H470, Z490',
    'Intel 500 Series': 'H510, B560, H570, Z590',
    'Intel 600 Series': 'H610, B650, H670, Z690',
    'Intel 700 Series': 'H710, B750, H770, Z790',
    'Intel 6 Series': 'H61, B65, H67, Z67',
    'Intel 7 Series': 'H71, B75, H77, Z77',
    'Intel 8 Series': 'H81, B85, H87, Z87',
    'Intel 9 Series': 'H91, B95, H97, Z97',
    'Intel 60 Series': 'H61, B65, H67, Z67',
    'Intel 70 Series': 'H71, B75, H77, Z77',
    'Intel 80 Series': 'H81, B85, H87, Z87',
    'Intel 90 Series': 'H91, B95, H97, Z97',
    'Intel H57': 'H57, ',
    'Intel H55': 'H55, ',
    'Intel P55': 'P55, ',
}
```

```python
def replace_series_names(series):
    if pd.isna(series) or series == '':
        return series
    
    parts = series.split(', ')
    replaced_parts = [remap.get(part, part) for part in parts]
    return ', '.join(replaced_parts)

df_normalized['supported_chipset'] = df_normalized['supported_chipset'].apply(replace_series_names)
df_normalized
```

```python
df_normalized['blenderPred'] = df_normalized['blenderPred'].apply(lambda x: x <= 0)
df_normalized
```

```python
df_indexed = df_normalized
df_indexed.index = range(1, len(df_indexed) + 1)
df_indexed
```

```python
# Save the DataFrame to a CSV file
df_indexed.to_csv('finalCPUlist.csv', index=True)

# Save the DataFrame to a JSON file
df_indexed.to_json('finalCPUlist.json', orient='records', indent=4)
```

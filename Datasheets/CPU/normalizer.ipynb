```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Projects\0FYP\Datasheets\CPU\Final Merger\CPU.csv')
df.head()
```

```python
df[['core_count_spec', 'boost_clock_spec', 'cache_spec', 'tdp_spec']] = df[['core_count', 'boost_clock', 'cache', 'tdp']]

df
```

```python
scaler_min_max = MinMaxScaler()
df[['core_count', 'boost_clock', 'cache', 'tdp']] = scaler_min_max.fit_transform(df[['core_count', 'boost_clock', 'cache', 'tdp']])
df
```

```python
df['tdp'] = 1 - df['tdp']
df
```

```python
df.to_csv('cpu_final.csv', index=False)
```

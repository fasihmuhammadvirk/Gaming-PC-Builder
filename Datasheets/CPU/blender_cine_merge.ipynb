```python
import pandas as pd
```

```python
main = pd.read_csv('cpupartsv3.csv')
main.head()
```

```python
main['boost_clock'] = main['boost_clock'].fillna(main['core_clock'])
main
```

```python
cinebench = pd.read_csv('cinebench_r15.csv')
cinebench = cinebench.rename(columns={'match': 'name'})
cinebench = cinebench.drop(columns={'Cores', 'GHz'})
cinebench
```

# Inner Merge CPU - Cinebench

```python
cpu_cine_IM = pd.merge(main, cinebench, on='name', how='inner')
#cpu_cine_IM.to_csv('cpu_cine_IM.csv', index=False)
```

```python
blender = pd.read_csv('blender.csv')
blender
```

```python
blender['name'] = blender['name'].str.replace(' PRO ', ' ')
blender['Blender Score'] = blender['Blender Score'].round(2)
blender
```

```python
blender_sorted = blender.sort_values(by='Number of Benchmarks', ascending=False)
blender_unique = blender_sorted.drop_duplicates(subset=['name'], keep='first')
#blender_unique.to_csv('blender - cleaned.csv')
```

# Inner Merge CPU - Blender

```python
cpu_blend_IM = pd.merge(main, blender_unique, on='name', how='inner')
#cpu_blend_IM.to_csv('cpu_blend_IM.csv', index=False)
```

```python
cpu_blend_IM
```

```python
passmark = pd.read_csv('passmark_cleaned.csv')
passmark.head()
```

```python
passmark = passmark.drop(columns=['TDP', 'cores', 'testDate', 'socket', 'cpuValue', 'threadValue', 'powerPerf'])
passmark = passmark.rename(columns={'cpuName': 'name', 'price': 'pmprice'})
passmark.head()
```

```python
cpu_pass_IM = pd.merge(main, passmark, on='name', how='inner')
cpu_pass_IM = cpu_pass_IM.drop_duplicates(subset=['name'], keep='last')
#cpu_pass_IM.to_csv('cpu_pass_IM.csv', index=False)
```

```python
cpu_pass_LM = pd.merge(main, passmark, on='name', how='left')
cpu_pass_LM = cpu_pass_LM.drop_duplicates(subset=['name'], keep='last')
cpu_pass_LM.to_csv('cpu_pass_LM.csv', index=False)
```

```python
cpu_pass_IM
```

```python
cpu_pass_blender_IM = pd.merge(cpu_pass_IM, blender_unique, on='name', how='inner')
cpu_pass_blender_IM
```

```python
cpu_pass_cine_IM = pd.merge(cpu_pass_IM, cinebench, on='name', how='inner')
cpu_pass_cine_IM
```

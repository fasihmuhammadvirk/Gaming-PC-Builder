```python
import pandas as pd
```

```python
api = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Imputations\gpu_api_impute.csv')
blender = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Imputations\gpu_blender_impute.csv')
fps = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Imputations\gpu_fps_impute.csv')
g3d = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Imputations\gpu_g3d_impute.csv')
```

```python
g3d
```

```python
cols_to_drop = ['core_clock', 'boost_clock', 'memory_clock', 'memory_ver', 'bus_width', 'shaders', 'TMUs', 'ROPs', 'PCIe', 'TDP', 'rec_psu', 'detail_URL']
api = api.drop(columns=cols_to_drop)
blender = blender.drop(columns=cols_to_drop)
fps = fps.drop(columns=cols_to_drop)
```

```python
api
```

```python
merged = pd.merge(g3d, api, how='inner', on=['chipset', 'memory'])
merged = pd.merge(merged, blender, how='inner', on=['chipset', 'memory'])
merged = pd.merge(merged, fps, how='inner', on=['chipset', 'memory'])
merged
```

```python
merged = merged.rename(columns={
    'G3Dmark': 'g3dmark', 
    'G2Dmark': 'g2dmark',	
    'Median Score': 'blender_score', 
    'Number of Benchmarks': 'no_benchmark', 
    '1080p perf (FPS)': '1080p_fps', 
    '1440p perf (FPS)': '1440p_fps', 
    '4k perf (FPS)': '4k_fps', 
    'overall FPS performance': 'fps_score'})

merged
```

```python
from sklearn.preprocessing import MinMaxScaler
df_normalized = merged
columns_to_normalize = ['g3dmark', 'g2dmark', 'api_score', 'blender_score', 'fps_score']

scaler = MinMaxScaler()
df_normalized[columns_to_normalize] = scaler.fit_transform(df_normalized[columns_to_normalize])
df_normalized
```

```python
df_normalized.to_csv('GPU.csv', index=False)
```

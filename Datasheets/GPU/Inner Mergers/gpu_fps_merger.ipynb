```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Projects\FYP prototype\ver 1.0\GPU\GPU fps benchmarks.csv")
df
```

```python
fps_col = ['1080p perf (FPS)', '1440p perf (FPS)', '4k perf (FPS)']

df['overall FPS performance'] = df[fps_col].mean(axis=1).round(2)
df = df.rename(columns={'Name': 'chipset'})
df
```

```python
df['chipset'] = df['chipset'].str.replace('NVIDIA ', '')
df['chipset'] = df['chipset'].str.replace('AMD ', '')
df['chipset'] = df['chipset'].str.replace('ATI ', '')
df
```

```python
base = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\gpu_base_main.csv')
base
```

```python
#base_stripped = base.drop(columns=['name', 'price', 'color', 'length'])
base_stripped = base.drop_duplicates(subset=['chipset', 'memory'])
base_stripped
```

```python
df = df.drop(columns=['Recommended PSU (in W)', 'Year', 'Max-Temp (in C)', 'Fan Noise (in dB)', 'V-RAM (in GB)'])
df = df.dropna(subset='overall FPS performance')
df
```

```python
base_fps_IM = pd.merge(base_stripped, df, on='chipset', how='inner')
base_fps_IM
```

```python
base_fps_IM.to_csv('base_fps_IM.csv', index=False)
```

```python
import pandas as pd
```

```python
base = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\gpu_base_main.csv')
base
```

```python
g3d = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\g3dmark_gpu.csv')
g3d = g3d.rename(columns={'gpuName': 'chipset'})
g3d
```

```python
#base_stripped = base.drop(columns=['name', 'price', 'color', 'length'])
base_stripped = base.drop_duplicates(subset=['chipset', 'memory'])
base_stripped
```

```python
g3d_stripped = g3d.drop(columns=['gpuValue', 'TDP', 'testDate', 'powerPerformance', 'category', 'price'])
g3d_stripped
```

```python
base_g3d_IM = pd.merge(base_stripped, g3d_stripped, on='chipset', how='inner')
base_g3d_IM
```

```python
base_g3d_IM.to_csv('base_g3d_IM.csv', index=False)
```

```python
import pandas as pd
```

```python
base = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\gpu_base_main.csv')
base.head()
```

```python
gpu_api = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\gpu_api_scores.csv')
gpu_api
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(1, 1000))

# Normalize the values in column 'A'
gpu_api['CUDA'] = scaler.fit_transform(gpu_api[['CUDA']])
gpu_api['Metal'] = scaler.fit_transform(gpu_api[['Metal']])
gpu_api['OpenCL'] = scaler.fit_transform(gpu_api[['OpenCL']])
gpu_api['Vulkan'] = scaler.fit_transform(gpu_api[['Vulkan']])
gpu_api
```

```python
gpu_api['api_score'] = gpu_api[['CUDA', 'Metal', 'OpenCL', 'Vulkan']].mean(axis=1).round(2)
gpu_api = gpu_api.rename(columns={'Device': 'chipset'}).drop(columns='Manufacturer')
gpu_api
```

```python
gpu_api = gpu_api.drop(columns=['CUDA', 'Metal', 'OpenCL', 'Vulkan'])
gpu_api
```

```python
#base_stripped = base.drop(columns=['name', 'price', 'color', 'length'])
base_stripped = base.drop_duplicates(subset=['chipset', 'memory'])
base_stripped
```

```python
base_api_IM = pd.merge(base_stripped, gpu_api, on='chipset', how='inner')
base_api_IM
```

```python
base_api_IM.to_csv('base_api_IM.csv', index=False)
```

```python
import pandas as pd
```

```python
base = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\gpu_base_main.csv')
base.head()
```

```python
blender = pd.read_csv(r'D:\Projects\FYP prototype\ver 1.0\GPU\blender_gpu.csv')
#blender = blender.rename(columns={'gpuName': 'chipset'})
blender
```

```python
#base_stripped = base.drop(columns=['name', 'price', 'color', 'length'])
base_stripped = base.drop_duplicates(subset=['chipset', 'memory'])
base_stripped
```

```python
blender = blender.rename(columns={'Device Name': 'chipset'})
blender['chipset'] = blender['chipset'].str.replace('NVIDIA ', '')
blender['chipset'] = blender['chipset'].str.replace('AMD ', '')
blender['chipset'] = blender['chipset'].str.replace('Apple ', '')
blender['chipset'] = blender['chipset'].str.replace('Intel ', '')
blender
```

```python
blender.to_csv('test.csv', index=False)
```

```python
base_blender_IM = pd.merge(base_stripped, blender, on='chipset', how='inner')
base_blender_IM
```

```python
import pandas as pd
import pickle
import numpy as np
```

```python
gpu_g3d_LM = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Left Mergers\base_g3d_LM.csv')
```

```python
with open('g3d_predictor.pkl', 'rb') as file:
    model = pickle.load(file)

input_columns = ['boost_clock', 'memory', 'memory_ver', 'bus_width', 'shaders', 'TMUs', 'ROPs', 'TDP']

df_new = gpu_g3d_LM
for index, row in df_new.iterrows():
    if np.isnan(row['G3Dmark']):
        X_pred = row[input_columns].values.reshape(1, -1)
        y_pred = model.predict(X_pred)      # making the prediction
        df_new.at[index, 'G3Dmark'] = float(y_pred[0][0])
        df_new.at[index, 'G2Dmark'] = float(y_pred[0][1])
        #df_new.at[index, 'passmarkPred'] = 'TRUE'
    #else:
        #df_new.at[index, 'passmarkPred'] = 'FALSE'
```

```python
df_new.to_csv('gpu_g3d_impute.csv', index=False)
```

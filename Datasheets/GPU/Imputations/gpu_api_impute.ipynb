```python
import pandas as pd
import pickle
import numpy as np
```

```python
gpu_api_LM = pd.read_csv(r'D:\Projects\0FYP\Datasheets\GPU\Left Mergers\base_api_LM.csv')
gpu_api_LM
```

```python
with open('api_predictor.pkl', 'rb') as file:
    model = pickle.load(file)

input_columns = ['boost_clock', 'memory', 'memory_ver', 'bus_width', 'shaders', 'TMUs', 'ROPs']

df_new = gpu_api_LM
for index, row in df_new.iterrows():
    if np.isnan(row['api_score']):
        X_pred = row[input_columns].values.reshape(1, -1)
        y_pred = model.predict(X_pred)      # making the prediction
        df_new.at[index, 'api_score'] = float(y_pred)
        #df_new.at[index, 'G2Dmark'] = float(y_pred[0][1])
        #df_new.at[index, 'Number of Benchmarks'] = 0
    #else:
        #df_new.at[index, 'passmarkPred'] = 'FALSE'
```
